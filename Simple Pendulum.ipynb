```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def pendulum(state,time): #the EOM for damped pendulum
    
    g0 = state[1]
    g1 = -g/l * np.sin(state[0]) - beta*state[1]
    
    return np.array([g0,g1])
```

```python

theta0 = 0.6  #initial conditions
thetadot0 = 0

N = 1000 #Number of time steps

l = 5   #parameters
g = 9.8
beta = 0.0 #without damping

time = np.linspace(0,40,N)

y = np.zeros(2)

y[0] = theta0
y[1] = thetadot0


answer = odeint(pendulum,y,time) #this solves the differential equations
```

```python

fig, axs = plt.subplots(2,figsize=(10,5),sharex=True)
fig.suptitle('Position and Velocities for undamped Simple Pendulum')
axs[0].plot(time,answer[:,0])
axs[0].set_ylabel('Position')
axs[1].set_ylabel('Velocity')
axs[1].set_xlabel('time')
axs[1].plot(time,answer[:,1])
plt.savefig("simplependulum.eps")
```

```python
#damping case

theta0 = 0.6
thetadot0 = 0

N = 1000

l = 5
g = 9.8
beta = 0.5

time = np.linspace(0,40,N)

y = np.zeros(2)

y[0] = theta0
y[1] = thetadot0


answer = odeint(pendulum,y,time)
```

```python

fig, axs = plt.subplots(2,figsize=(10,5),sharex=True)
fig.suptitle('Position and Velocities for damped Simple Pendulum')
axs[0].plot(time,answer[:,0])
axs[0].set_ylabel('Position')
axs[1].set_ylabel('Velocity')
axs[1].set_xlabel('time')
axs[1].plot(time,answer[:,1])
plt.savefig("dampedsimplependulum.eps")
```

```python
def dampdrivpendulum(state,time): #with a driving frorce
    
    g0 = state[1]
    g1 = -g/l * np.sin(state[0]) - beta*state[1] + A*np.cos(w*time)
    
    return np.array([g0,g1])
```

```python
theta0 = -4
thetadot0 = 0

N = 1000

l = 6
g = 9.8
beta = 0.5

time = np.linspace(0,100,N)
w = 0.9
y = np.zeros(2)
A = 1.5

y[0] = theta0
y[1] = thetadot0


answer = odeint(dampdrivpendulum,y,time)
```

```python

fig, axs = plt.subplots(2,figsize=(10,5),sharex=True)
fig.suptitle('Position and Velocities for driven damped Simple Pendulum')
axs[0].plot(time,answer[:,0])
axs[0].set_ylabel('Position')
axs[1].set_ylabel('Velocity')
axs[1].set_xlabel('time')
axs[1].plot(time,answer[:,1])
plt.savefig("dampeddrivensimplependulum.eps")
```
