```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pendulum(state,time):
    
    g0 = state[1]
    g1 = -g/l * np.sin(state[0]) - beta*state[1]
    
    return np.array([g0,g1])
```

```python
from scipy.integrate import odeint
```

```python

theta0 = 0.6
thetadot0 = 0

N = 1000

l = 5
g = 9.8
beta = 0.0

time = np.linspace(0,40,N)

y = np.zeros(2)

y[0] = theta0
y[1] = thetadot0


answer = odeint(pendulum,y,time)
```

```python
plt.plot(time,answer[:,0])
```

```python
def dampdrivpendulum(state,time):
    
    g0 = state[1]
    g1 = -g/l * np.sin(state[0]) - beta*state[1] + A*np.cos(w*time)
    
    return np.array([g0,g1])
```

```python

theta0 = -4
thetadot0 = 0

N = 1000

l = 6
g = 9.8
beta = 0.5

time = np.linspace(0,100,N)
w = 0.9
y = np.zeros(2)
A = 1.5

y[0] = theta0
y[1] = thetadot0


answer = odeint(dampdrivpendulum,y,time)
```

```python
plt.plot(time,answer[:,0])
```
